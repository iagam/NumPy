## Plotting with NumPy and Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plot a straight line by placing two munpy arrays of the same length

plt.plot(np.arange(10), np.arange(10))
plt.show()
```

```python
# Plot an exponential

plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.show()
```

```python
# Place the two graphs side by side

plt.figure()
plt.subplot(121)
plt.plot(np.arange(10), np.arange(10))
plt.subplot(122)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.show()
```

```python
# Place the two graphs top to bottom

plt.figure()
plt.subplot(211)
plt.plot(np.arange(10), np.arange(10))
plt.subplot(212)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.show()
```

```python
# plot a 2x2 grid traversing in reading order from one to four

plt.figure()
plt.subplot(221)
plt.plot(np.arange(10), np.arange(10))
plt.subplot(222)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(223)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(224)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
plt.show()
```
